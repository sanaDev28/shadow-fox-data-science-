```python
import matplotlib.pyplot as plt

subjects = ['Math', 'Science', 'English', 'Hindi']
marks = [85, 90, 78, 88]

colors = ['#6C5CE7', '#00B894', '#FDCB6E', '#E17055']

plt.figure(figsize=(8, 8))

wedges, texts, autotexts = plt.pie(
    marks,
    labels=subjects,
    colors=colors,
    autopct='%1.1f%%',
    startangle=140,
    explode=(0.03, 0.08, 0.03, 0.03),
    shadow=True,
    wedgeprops={'edgecolor': 'white', 'linewidth': 2},
    textprops={'fontsize': 12}
)
plt.title(
    'Subject Marks Distribution',
    fontsize=18,
    fontweight='bold',
    pad=20
)

plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
iris = load_iris()
x = iris.data[:, 0]
y = iris.data[:, 1]

plt.figure(figsize=(8, 6))

plt.scatter(
    x, y,
    s=80,
    alpha=0.7,
    edgecolors='black',
    color='green'
)

plt.title('Iris Dataset Scatter Plot', fontsize=16, fontweight='bold')
plt.xlabel('Sepal Length (cm)', fontsize=12)
plt.ylabel('Sepal Width (cm)', fontsize=12)
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt

marks = [45, 52, 55, 60, 62, 65, 68, 70, 72, 75,
         78, 80, 82, 85, 87, 90, 92, 95, 98, 100]

plt.figure(figsize=(8, 5))
plt.hist(
    marks,
    bins=6,
    color='orange',
    edgecolor='black',
    linewidth=1.5
)
plt.title('Student Marks Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Marks', fontsize=12)
plt.ylabel('Number of Students', fontsize=12)
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
sns.set_style("darkgrid")
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", palette="Set2", s=100)
plt.title("Total Bill vs Tip", fontsize=14)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
marks = [
    [85, 90, 78, 92],
    [88, 76, 95, 89],
    [70, 85, 80, 75],
    [95, 92, 88, 90]
]
sns.heatmap(
    marks,
    annot=True,
    cmap="YlGnBu",
    linewidths=0.5
)
plt.title("Student Marks Heatmap")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
group = ["A", "A", "B", "B", "C", "C", "D", "D"]
scores = [65, 70, 75, 80, 85, 90, 92, 88]
sns.violinplot(x=group, y=scores, hue=group, palette="Set1", legend=False)
plt.title("Multiple Violin Plots")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
grade= ["A","A","B","B","C","C"]
student = [65, 70, 75, 80, 85, 90]
sns.boxplot(x=grade, y=student, palette="Set1",hue=grade,legend=False)
plt.title("Multiple Box Plots")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
grade = ["A", "B", "C", "D"]
student = [10, 25, 15, 30]
sns.barplot(x=grade, y=student, palette="Set1",hue=grade,legend=False)
plt.title("Simple Bar Plot")
plt.xlabel("Group")
plt.ylabel("Values")
plt.show()
```
